```python
def ages_dict(filename):
    newfile = filename
    with open(newfile, 'r') as ff:
        canid_recs = ff.readlines()
    from collections import defaultdict

    species_ages = defaultdict(list)

    for line in canid_recs:
        taxon = line.split(",")[6]
        if taxon == '"species"':
            species = line.split(",")[5]
            minage = line.split('","')[10]
            maxage = line.split('","')[11]
            mean_age = (float(minage) + float(maxage)) / 2
            species_ages[species].append(mean_age)
    return(species_ages)
```

```python
species_ages = ages_dict("/home/eeb177-student/Desktop/eeb-177/class-assignments/canids/cleaned_canids.csv")
```

```python
def range_file(filename, dictname):
    output = open(filename, "w")

    for key in dictname.keys():
        ages = dictname[key]
        minage = min(ages)
        maxage = max(ages)
        genus = key.replace('"','').split(" ")[0]
        species = key.replace('"','')
        outline = "{},{},{},{}\n".format(genus,species,minage,maxage)
        output.write(outline)
    output.close()
```

```python
ranges = range_file("/home/eeb177-student/Desktop/eeb-177/class-assignments/canids/canid-ranges.csv", species_ages)
```
